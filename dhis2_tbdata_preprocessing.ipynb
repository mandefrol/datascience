# Grouping the filtered data region by region

```python
# let us check if there is any empty row at the regional level
df_tb_all_forms_referred_by_community[df_tb_all_forms_referred_by_community.orgunitlevel2.isna()]
```

```python
# Let us start with grouping on regional basis
df_tb_allforms_referred_by_community_by_region = df_tb_all_forms_referred_by_community.groupby('orgunitlevel2').sum(numeric_only=True)
df_tb_all_forms_referred_by_ppm_by_region = df_tb_all_forms_referred_by_ppm.groupby('orgunitlevel2').sum(numeric_only=True)
df_tb_all_relapse_by_region = df_tb_all_relapse.groupby('orgunitlevel2').sum(numeric_only=True)
df_tb_allforms_treated_by_region = df_tb_allforms_treated.groupby('orgunitlevel2').sum(numeric_only=True)
```

```python
with pd.ExcelWriter('data/tb_all_forms_dhis2_regions.xlsx') as writer:
    df_tb_allforms_referred_by_community_by_region.to_excel(writer, sheet_name='allforms_bycommunity')
    df_tb_all_forms_referred_by_ppm_by_region.to_excel(writer, sheet_name='allforms_byppm')
    df_tb_all_relapse_by_region.to_excel(writer, sheet_name='all_relapse_detected')
    df_tb_allforms_treated_by_region.to_excel(writer, sheet_name='allforms_treated')
print('Successful')
```

```python
# Let us now group by the zonal/district level
df_tb_allforms_referred_by_community_by_district = df_tb_all_forms_referred_by_community.groupby(
    ['orgunitlevel2','orgunitlevel3']).sum(numeric_only=True)
df_tb_all_forms_referred_by_ppm_by_district = df_tb_all_forms_referred_by_ppm.groupby(
    ['orgunitlevel2','orgunitlevel3']).sum(numeric_only=True)
df_tb_all_relapse_by_district = df_tb_all_relapse.groupby(
    ['orgunitlevel2','orgunitlevel3']).sum(numeric_only=True)
df_tb_allforms_treated_by_district = df_tb_allforms_treated.groupby(
    ['orgunitlevel2','orgunitlevel3']).sum(numeric_only=True)
```

```python
# Export the district level data to excel file
with pd.ExcelWriter('data/tb_all_forms_dhis2_districsts.xlsx') as writer:
    df_tb_allforms_referred_by_community_by_district.to_excel(writer, sheet_name='allforms_bycommunity')
    df_tb_all_forms_referred_by_ppm_by_district.to_excel(writer, sheet_name='allforms_byppm')
    df_tb_all_relapse_by_district.to_excel(writer, sheet_name='all_relapse_detected')
    df_tb_allforms_treated_by_district.to_excel(writer, sheet_name='allforms_treated')
print('Successful')
```
