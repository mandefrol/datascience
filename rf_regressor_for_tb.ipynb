```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

```python
dataset = pd.read_csv('data/tb_detection_rate_eth_who.csv')
dataset['Year'] = list(range(2000,2022))
dataset.head()
```

```python
X = dataset.iloc[:, 0:-1].values
y = dataset.iloc[:, -1].values
```

```python
regressor = RandomForestRegressor(n_estimators=21, random_state=0)
regressor.fit(X, y)
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, regressor.predict(X), color='blue')
```

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
```

```python
X_forecast = np.arange(2022, 2031, 0.1)
X_forecast = X_forecast.reshape((len(X_forecast), 1))
plt.plot(X_forecast, regressor.predict(X_forecast))
```
